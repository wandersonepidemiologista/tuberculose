```python
# prompt: tabela com numero absoluto e percentual do campo CS_RACA

import pandas as pd

# Calcular a frequência absoluta de cada valor na coluna 'CS_RACA'
frequencia_raca = df['CS_RACA'].value_counts()

# Calcular a frequência relativa (percentual) de cada valor
percentual_raca = (df['CS_RACA'].value_counts(normalize=True) * 100).round(2)

# Criar um DataFrame com as frequências absolutas e relativas
tabela_raca = pd.DataFrame({'Frequência Absoluta': frequencia_raca, 'Frequência Relativa (%)': percentual_raca})

# Exibir a tabela
print(tabela_raca)
```

```python
# prompt: no campo CS_RACA faça a substituição dos valores para '1.0' : 'Branca', '2.0' : 'Preta',  '3.0' : 'Amarela',  '4.0' : 'Parda',  '5.0' : 'Indígena',  '9.0' : 'Ignorada'

raca_mapping = {
    1.0: 'Branca',
    2.0: 'Preta',
    3.0: 'Amarela',
    4.0: 'Parda',
    5.0: 'Indígena',
    9.0: 'Ignorada'
}

df['CS_RACA_DESCRICAO'] = df['CS_RACA'].map(raca_mapping)

# Print the updated DataFrame to verify the changes
print(df[['CS_RACA', 'CS_RACA_DESCRICAO']].head(20))
```
